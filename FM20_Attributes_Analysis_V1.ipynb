```python
# Import required packages
import numpy as np
import pandas as pd
#import plotly.express as px
from easygui import *
import matplotlib.pyplot as plt
```

```python
# Choose data to analyse : Current Squad or Potential Signings
choice = buttonbox ("What data would you like to analyse?", title = "Data source selection", choices = ["Current Squad", "Potential Signings"])
if choice == "Current Squad":
    df = pd.read_html(r"C:\Users\stuar\Documents\Sports Interactive\Football Manager 2020\html\attributes_view.html",skiprows=0,)[0]
else:
    df = pd.read_html(r"C:\Users\stuar\Documents\Sports Interactive\Football Manager 2020\html\scouting_attributes.html",skiprows=0)[0]
```

```python
# Set DataFrame limits
pd.set_option('display.max_columns',50)
pd.set_option('display.max_rows',5)
pd.set_option('precision', 0)
```

```python
# Create score for each position and club DNA.
df['Club DNA'] = df['Agg'] + df['Det'] + df['Tea'] + df['Wor']
df['Sweeper Keeper'] = df['Com'] + df['Kic'] + df['1v1'] + df['Ref'] + df['TRO'] + df['Ant'] + df['Cmp'] + df['Cnt'] + df['Pos'] + df['Agi']
df['Complete Wing Back'] = df['Cro'] + df['Dri'] + df['Fir'] + df['Pas'] + df['Tec'] + df['Dec'] + df['OtB'] + df['Tea'] + df['Wor'] + df['Acc'] + df['Pac'] + df['Sta']
df['Wing Back'] = df['Cro'] + df['Dri'] + df['Mar'] + df['Tck'] + df['OtB'] + df['Tea'] + df['Wor'] + df['Acc'] + df['Sta'] 
df['Central Defender'] = df['Hea'] + df['Mar'] + df['Tck'] + df['Pos'] + df['Jum'] + df['Str'] 
df['Defensive Midfielder'] = df['Tck'] + df['Ant'] + df['Cnt'] + df['Pos'] + df['Tea'] + df['Wor'] + df['Sta'] 
df['Ball Winning Midfielder'] = df['Tck'] + df['Agg'] + df['Ant'] + df['Bra'] + df['Tea'] + df['Wor'] 
df['Carillero'] = df['Pas'] + df['Tck'] + df['Dec'] + df['Pos'] + df['Tea'] + df['Wor'] + df['Sta'] 
df['Inside Forward Support'] = df['Dri'] + df['Fir'] + df['Pas'] + df['Tec'] + df['OtB'] + df['Acc'] + df['Agi'] + df['Bal'] 
df['Inside Forward Attack'] = df['Dri'] + df['Fin'] + df['Fir'] + df['Tec'] + df['OtB'] + df['Acc'] + df['Agi'] + df['Bal'] 
df['False Nine'] = df['Dri'] + df['Fir'] + df['Pas'] + df['Tec'] + df['Cmp'] + df['Dec'] + df['OtB'] + df['Vis'] + df['Acc'] + df['Agi']
```

```python
# Create custom column on DataFrame for each position and DNA
dna = df[['Name', 'Age', 'Agg' , 'Det' , 'Tea' , 'Wor', 'Club DNA']]
sk = df[['Name', 'Age' , 'Com' , 'Kic' , '1v1', 'Ref', 'TRO', 'Ant', 'Cmp', 'Cnt', 'Pos', 'Agi', 'Sweeper Keeper']]
cwb = df[['Name', 'Age' , 'Cro' , 'Dri', 'Fir', 'Pas', 'Tec', 'Dec', 'OtB', 'Tea', 'Wor', 'Acc', 'Pac', 'Sta', 'Complete Wing Back']]
wb = df[['Name', 'Age' , 'Cro' , 'Dri', 'Mar', 'Tck', 'OtB', 'Tea', 'Wor', 'Acc', 'Sta', 'Wing Back']] 
cd = df[['Name', 'Age' ,  'Hea', 'Mar' , 'Tck', 'Pos', 'Jum', 'Str' , 'Central Defender']] 
dm = df[['Name', 'Age' , 'Tck', 'Ant', 'Cnt', 'Pos', 'Tea', 'Wor', 'Sta', 'Defensive Midfielder']]
bwm = df[['Name', 'Age' , 'Tck' , 'Agg' , 'Ant' , 'Bra' , 'Tea' , 'Wor', 'Ball Winning Midfielder']]
car = df[['Name', 'Age' , 'Pas', 'Tck', 'Dec', 'Pos', 'Tea', 'Wor', 'Sta', 'Carillero']]
ifs = df[['Name', 'Age' , 'Dri', 'Fir', 'Pas', 'Tec', 'OtB', 'Acc', 'Agi', 'Bal', 'Inside Forward Support']]
ifa = df[['Name', 'Age' , 'Dri', 'Fin', 'Fir', 'Tec', 'OtB', 'Acc', 'Agi', 'Bal', 'Inside Forward Attack']]
f9 = df[['Name', 'Age' , 'Dri', 'Fir', 'Pas', 'Tec', 'Cmp', 'Dec', 'OtB', 'Vis', 'Acc', 'Agi', 'False Nine']]
```
